```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Confusion Matrix
confusion_matrix = [
    [1011, 0, 6, 1],
    [0, 93, 7, 0],
    [4, 4, 986, 18],
    [1, 0, 49, 918]
]

# Function to calculate accuracy for each class
def calculate_accuracy(matrix, class_index):
    TP = matrix[class_index][class_index]  # True Positive for the class
    FN = sum(matrix[class_index]) - TP    # False Negative for the class
    FP = sum([row[class_index] for row in matrix]) - TP  # False Positive for the class
    TN = sum([sum(row) for row in matrix]) - (TP + FN + FP)  # True Negative for the class
    accuracy = (TP + TN) / (TP + TN + FP + FN)  # Accuracy formula
    return accuracy

# Calculate accuracy for each class
accuracy_mild = calculate_accuracy(confusion_matrix, 0)
accuracy_moderate = calculate_accuracy(confusion_matrix, 1)
accuracy_non = calculate_accuracy(confusion_matrix, 2)
accuracy_very_mild = calculate_accuracy(confusion_matrix, 3)

# Define the data for the classification report with accuracy
data = {
    'mild': [1.00, 0.99, 1.0],
    'moderate': [1.00, 0.93, 0.96],
    'non': [0.97, 0.97, 0.94],
    'very_mild': [0.99, 0.95, 0.98],
}

# Create a DataFrame
classification_report_df = pd.DataFrame(data, index=['specificity', 'recall', 'precision'])

# Function to create a formatted figure (with title and figure number)
def create_figure(fig_num, title):
    plt.figtext(0.02, 1.07, fig_num, ha="left", va="top", fontsize=12, fontweight="bold")
    plt.figtext(0.02, 1.03, title, ha="left", va="top", fontsize=11, fontstyle="italic")

# Create the plot
fig, ax = plt.subplots(figsize=(10, 6))

# Create the figure number and italicized title
create_figure("Figure 3", "Classification Report for Scratch Model")

# Add another non-italicized title just above the heatmap

# Create the heatmap
sns.heatmap(classification_report_df, annot=True, fmt='.2f', cmap='Blues', cbar=True,
            xticklabels=classification_report_df.columns, yticklabels=classification_report_df.index)

# Add labels
plt.xlabel("Class", fontsize=12)
plt.ylabel("Metrics", fontsize=12)

# Adjust layout to ensure proper title and figure number placement
plt.subplots_adjust(top=0.85, bottom=0.15)  # Adjust top margin to make space for title

# Show the plot
plt.tight_layout()
plt.show()
```
